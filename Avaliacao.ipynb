```python
avaliacao_excelente = 0
avaliacao_ruim=0

for i in range(10):
  nome = str(input("Digite seu nome : "))
  idade = int(input("Digite sua idade: "))
  while idade >100 or idade <5: idade = int(input("Erro! A idade não pode ser maior que 100 anos ou menor que 05 anos. Digite novamente: "))

  avaliacao = int(input("Avalie o atendimento de 1 a 3. Sendo 1-Excelente, 2-Bom, 3-Ruim: "))
  while avaliacao <=0 or avaliacao >=4: avaliacao= int(input("Erro! O atendimento deverá ser avaliado apenas na escala de 1 a 3. Sendo 1-Excelente, 2-Bom, 3-Ruim: "))
  print("Obrigada pela avaliação!!")
  if avaliacao==1: avaliacao_excelente +=1
  elif avaliacao==3: avaliacao_ruim +=1

print('---Total de avaliações---')
print(f'Excelente: {avaliacao_excelente}')
print(f'Ruim: {avaliacao_ruim}')
```
