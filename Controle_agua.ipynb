```python
from colorama import Fore, Style

def nivel_agua():
    mensagem = [
        Fore.RED + "Nível 1: Muito baixo (crítico)",
        Fore.YELLOW + "Nível 2: Baixo",
        Fore.GREEN + "Nível 3: Médio",
        Fore.CYAN + "Nível 4: Alto",
        Fore.BLUE + "Nível 5: Muito alto (alerta)"
    ]

    while True:
        nivel = input("Digite o nível do reservatório (1 a 5): ")
        if nivel.isdigit():
          nivel_reservatorio = int(nivel)

          if 1 <= nivel_reservatorio <= 5:
              print(f"Status: {mensagem[nivel_reservatorio - 1]}")
              break
          else:
            print("Erro: Nível informado fora do intervalo (1-5).")
        else:

          print("Erro: Por favor, apenas números.")

nivel_agua()
```
