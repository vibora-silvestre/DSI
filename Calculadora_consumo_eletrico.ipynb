```python
nome_aparelho = input("Digite o nome do aparelho: ")
potencia_watts = float(input("Informe a potêcia em watts: "))
tempo_horas = float(input("Informe o tempo médio de uso diário do aparelho em horas: "))
while tempo_horas >=24 or tempo_horas <=-1: tempo_horas =float(input("Tempo diário não poderá ser maior que 24 horas e nem negativo, informe um tempo válido: "))

consumo_mensal = (potencia_watts * tempo_horas * 30) / 1000
consumo_mensal_dinheiro = float(consumo_mensal * 0.85)
print("\n \n ----Dados de consumo---- \n Aparelho:", nome_aparelho, "Seu consumo mensal foi:",consumo_mensal, "KwH/Mês, resultando em um gasto de R$:", round(consumo_mensal_dinheiro,2))
```
